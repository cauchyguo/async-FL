```python
import torch
from torchvision import datasets, transforms
import matplotlib.pyplot as plt

# Tensor
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])

#  FashionMNIST 
train_dataset = datasets.FashionMNIST(root='/home/ypguo/async-FL/src/data/', train=True, download=True, transform=transform)

# 10
unique_labels = set(train_dataset.targets.numpy())  # 
fig, axes = plt.subplots(2, 5, figsize=(10, 5))

for i, label in enumerate(unique_labels):
    # 
    idx = (train_dataset.targets == label).nonzero(as_tuple=True)[0][0]
    image, target = train_dataset[idx]

    # 
    ax = axes[i // 5, i % 5]
    ax.imshow(image.squeeze(), cmap='gray')
    ax.set_title(f"Label: {label}")
    ax.axis('off')

plt.tight_layout()
plt.show()

import torch
from torchvision import datasets, transforms
import matplotlib.pyplot as plt

# Tensor
transform = transforms.Compose([transforms.ToTensor(), ])

#  CIFAR-10 
train_dataset = datasets.CIFAR10(root='/home/ypguo/async-FL/src/data/', train=True, download=False, transform=transform)

# targetsTensor
targets_tensor = torch.tensor(train_dataset.targets)

# 
unique_labels = set(targets_tensor.numpy())  # 
fig, axes = plt.subplots(2, 5, figsize=(10, 5))

for i, label in enumerate(unique_labels):
    # 
    idx = (targets_tensor == label).nonzero(as_tuple=True)[0][0]
    image, target = train_dataset[idx]

    # 
    ax = axes[i // 5, i % 5]
    ax.imshow(image.permute(1, 2, 0))  #  (C, H, W) -> (H, W, C)
    ax.set_title(f"Label: {label}")
    ax.axis('off')

plt.tight_layout()
plt.show()

```
